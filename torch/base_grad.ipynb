```python
import torch
```

```python
x = torch.ones(2, requires_grad=True)
print(x, x.shape, x.size())
z = x + 2
print(z, z.shape, z.size())
z.backward(torch.ones_like(z))
# z.backward() # Error: grad can be implicitly created only for scalar outputs
print(x.grad)
```

```python
# 不清除梯度张量，只把梯度值归零
x.grad.zero_()
# x.grad = None  # 清除梯度张量
m = x + 2
m.sum().backward()
print(x.grad)
```

```python
x = torch.tensor([2., 1.], requires_grad=True)
y = torch.tensor([[1., 2.], [3., 4.]], requires_grad=True)

z = torch.mm(x.view(1, 2), y)
print(f"z:{z}")
z.backward(torch.Tensor([[1., 0]]), retain_graph=True)
print(f"x.grad: {x.grad}")
print(f"y.grad: {y.grad}")
```
