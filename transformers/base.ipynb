# Load Base Model
load model from huggingface hub(.cache) and save to other directory

```python
from transformers import AutoTokenizer, AutoModel
import os

# Define the model name and local directory
model_name = "bert-base-chinese"  # You can change this to any model you need
local_model_dir = "/data/home/zhangchangtian/data/BaseModels/bert-base-chinese"

# Create directory if it doesn't exist
os.makedirs(local_model_dir, exist_ok=True)

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Save to local directory
# model_path = os.path.join(local_model_dir, model_name)
model_path = local_model_dir
tokenizer.save_pretrained(model_path)

# 如果要保存为pytorch_model.bin
# model.save_pretrained(model_path, safe_serialization=False)
model.save_pretrained(model_path)

print(f"Model and tokenizer saved to {model_path}")
```
