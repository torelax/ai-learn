```python
# embedding_params: vocab_size * hidden_size
emd_params = 30522 * 768
pos_params = 512 * 768 # 单子句子最大长度512
token_type_params = 2 * 768
emb_layer_norm_weights_params =  768
emb_layer_norm_bias_params = 768

embedding_params = emd_params + pos_params + token_type_params + emb_layer_norm_weights_params + emb_layer_norm_bias_params
print('Embedding module Params: ', embedding_params)
```

```python
# encoder params
attn_params = 12 * (768 * 768 * 3 + 768 * 3)  # weight of q, k, v and bias
attn_output_params = 12 * (768 * 768 + 768)  # num_attention_heads * (hidden_size * hidden_size + hidden_size)
layer_norm = 12 * 2 * (768 + 768)  # 2 * (weight + bias) for each layer
encoder_layer_params = attn_params + attn_output_params + layer_norm
print('Attention and LayerNorm Params: ', encoder_layer_params)
```

```python
# FFN params
# head_num * (hidden_size * intermediate_size + bias(intermediate_size) + intermediate_size * hidden_size + bias(hidden_size))
ffn_params = 12 * (768 * 3072 + 3072 + 3072 * 768 + 768)
print('Feed-Forward Network Params: ', ffn_params)
```

```python
# pooler params
pooler_params = 768 * 768 + 768  # weight + bias
print('Pooler Params: ', pooler_params)
```

```python
print('Total Params: ', embedding_params + encoder_layer_params + ffn_params + pooler_params)
```
