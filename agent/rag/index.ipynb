# split doc with langchain

```python
from langchain_text_splitters.base import TextSplitter, Document

s = Document(page_content="This is a test. This is only a test.", metadata={"source": "test.txt"})
tp = TextSplitter()

tp.create_documents(texts=["This is a test. This is only a test."])
tp.split_documents(text="This is a test. This is only a test.")
```
